```python
#with statement - automatically close krnw file eka read krla passe
#advantage - file eka close krnw naththam wadi memory use krnw na , error ekk awoth file eka close krnw na
with open("sample_1.txt", "r") as file_2:
    content_2 = file_2.read()
    print(content_2)
```

```python
#readline() - read one line at a time
with open("sample_1.txt", "r") as file_3:
    line_1 = file_3.readline()
    print(line_1)
    line_2 = file_3.readline()
    print(line_2)
    line_3 = file_3.readline()
    print(line_3)
```

```python
with open("sample_1.txt", "r") as file_3:
    line_1 = file_3.readline()
    line_1 = line_1.strip()
    print(line_1)
    line_2 = file_3.readline()
    print(line_2)
    line_3 = file_3.readline()
    print(line_3)
```

```python
#file eke okkom content ek ek parin readline() eken print krnn
with open ("sample_1.txt", "r") as file_4:
    while True:
        file_4_content = file_4.readline()
        if not file_4_content:
            break
        print(file_4_content.strip())
```

```python
#readlines() - read all lines and return a list of lines
with open("sample_1.txt", "r") as file_5:
    lines = file_5.readlines()
    print(lines)
```

```python
with open("sample_1.txt", "r") as file_6:
    lines = file_6.readlines()
    for line in lines:
        print(line.strip())
```

```python
#nextLine() - read next line
with open("sample_1.txt", "r") as file_4:
    for _ in range(3):
        line = next(file_4)
        print(line)
```

```python
#DictReader - read csv file data as dictionary
#each row in the csv file will be represented as a dictionary where the keys are the column headers
import csv

with open("person_data.csv", "r") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        print(row)
```

```python
#csv file ekt data add krnne dictionary wlin smnwith list ekk hdagnn 

import csv

csv_file_path = "person_data.csv"

new_rows = [
        {"Name": "John", "Age": 28, "City": "New York"},
        {"Name": "Emma", "Age": 22, "City": "Los Angeles"},
        {"Name": "Michael", "Age": 35, "City": "Chicago"}
    ]

fieldnames = ["Name", "Age", "City"]



with open(csv_file_path, "w", newline='') as csv_file:
 
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()
    csv_writer.writerows(new_rows)

print(f"csv file updated successfully: {csv_file_path}")



    
```

```python
#write a python code that reads a file called  'input.text' and writes only the even-numbered lines into a new file called 'even_lines.txt' .(consider the first line as line number 1)
with open('input.txt', 'r') as file_2:
    file_2_reader = file_2.read()
print(file_2_reader)

with open('input.txt', 'r') as _f:
    lines = _f.readlines()

even_lines = []
with open('even_line.txt', 'w') as file_2:
    count = 0
    for line in lines:
        count += 1
        if count % 2 == 0:
            file_2.write(line)

print(even_lines)
                 
```
