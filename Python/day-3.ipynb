```python
# Given a list of numbers:
# - if it has 2 numbers, treat them as base and height of a right-angled triangle and compute the hypotenuse
# - if it has 3 numbers, treat them as triangle sides and compute the area using Heron's formula
# This implementation validates inputs and demonstrates usage with examples.
import math

def process_numbers(numbers):
    if not isinstance(numbers, (list, tuple)):
        print("Invalid input: expected a list or tuple of numbers")
        return

    if len(numbers) == 2:
        base, height = numbers
        try:
            base = float(base)
            height = float(height)
        except Exception:
            print("Invalid numbers for base/height")
            return
        if base > 0 and height > 0:
            hypotenuse = math.hypot(base, height)
            print(f"Numbers: {numbers} -> Hypotenuse: {hypotenuse}")
        else:
            print("Base and height must be positive numbers")

    elif len(numbers) == 3:
        a, b, c = numbers
        try:
            a = float(a); b = float(b); c = float(c)
        except Exception:
            print("Invalid numbers for triangle sides")
            return
        if a <= 0 or b <= 0 or c <= 0:
            print("Triangle sides must be positive numbers")
            return
        # Check triangle inequality
        if a + b <= c or a + c <= b or b + c <= a:
            print(f"Not a valid triangle: sides {numbers}")
            return
        s = (a + b + c) / 2
        area = math.sqrt(max(0.0, s * (s - a) * (s - b) * (s - c)))
        print(f"Numbers: {numbers} -> Area: {area}")

    else:
        print("Invalid input: expected a list with 2 or 3 numbers")


# Examples / demonstration
examples = [
    [3, 4],        # right triangle (3-4-5)
    [4, 7, 2],     # three sides -> area (will validate triangle inequality)
    [1, -2],       # invalid (negative)
    [1, 2, 3, 4],  # invalid length
]

for ex in examples:
    process_numbers(ex)
```
