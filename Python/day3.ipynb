```python
# using for loop factorial of a number and validate input
num = int(input("Enter a number: "))
factorial = 1
if num < 0:
    print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, num + 1):
        factorial = factorial * i
    print("The factorial of", num, "is", factorial)
```

```python
n = 5  # star pattern
for i in range(n):
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

```python
n = 5  # star pattern in one loop
for i in range(n):
    print("* " * (i + 1))
```

```python
total = 0
for i in range(10):
    num = int(input("Enter an integer (0 to stop): "))
    if num == 0:
        break
    total += num
print("The sum of all numbers entered before 0 is:", total)
```
