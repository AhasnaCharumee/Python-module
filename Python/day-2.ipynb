```python
sports = ("Soccer", "Basketball", "Baseball", "Tennis")
print(sports)
print(type(sports))
print(sports[1])
```

```python
sports[1] = "Cricket"  # This will raise an error because tuples are immutable
print(sports)
```

```python
sports = ("Soccer",)
print(type(sports))  # This will show <class 'tuple'> because it's a tuple
```
