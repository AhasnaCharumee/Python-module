```python
class Square:
    def _init_(self, name):
        self.__name = name

    def __area(self, side_length):
        return side_length ** 2
    
    def get_name(self):
        return self.__name

    def get_area(self, side_length):
        return self.__area(side_length) 
    
    def set_area(self, side_length):
        return self.__area(side_length)

square_1 = Square("square")
print(square_1.get_name())
print(square_1.get_area(5))
print(square_1.set_area(10))
```
