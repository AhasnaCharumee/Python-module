```python
#Function ‡∂ë‡∂ö: calculate_total(price, quantity = 1, discount = 0)

#‡∂∏‡∑ô‡∂∏ function ‡∂ë‡∂ö‡∑ö parameters:

#price ‚Üí ‡∂ë‡∂ö‡∂ö‡∂ß ‡∂á‡∂≠‡∑í ‡∑Ä‡∂ß‡∑í‡∂±‡∑è‡∂ö‡∂∏

#quantity (default = 1) ‚Üí ‡∂ú‡∂±‡∑ä‡∂± ‡∂¥‡∑ä‚Äç‡∂ª‡∂∏‡∑è‡∂´‡∂∫

#discount (default = 0) ‚Üí ‡∂Ö‡∂©‡∑î‡∂ö‡∑í‡∂ª‡∑ì‡∂∏ (rupees)

def calculate_total(price, quantity=1, discount=0):
    subtotal = price * quantity
    total = subtotal - discount

    if total != 0:
        discount_percentage = (discount / subtotal) * 100
    else:
        discount_percentage = 0
    return total, discount_percentage

total,discount_percentage = calculate_total(100, 1, 20)
print("total price is:", total)
print("discount percentage is:", discount_percentage, "%")


```

```python
def average_positive(*numbers):
    total=  0
    count = 0
    for number in numbers:
        if number > 0:
            total += number
            count += 1
        else:
            print("Ignoring non-positive number:", number)
    avg = total / count
    print("The average of positive numbers is:", avg)
    return avg

average_positive(10, -5, 20, 0, 15, -3, 30)

#average_positive() ‡∂ö‡∑í‡∂∫‡∂± function ‡∂ë‡∂ö ‡∂ë‡∂∫‡∂ß pass ‡∂ö‡∂ª‡∂± numbers ‡∂ë‡∂ö‡∑ô‡∂±‡∑ä positive (0 ‡∂ß ‡∑Ä‡∑ê‡∂©‡∑í) numbers ‡∑Ä‡∑í‡∂≠‡∂ª ‡∑Ä‡∑ê‡∂Ω‡∑í‡∂ú‡∑ô‡∂± average ‡∂ë‡∂ö ‡∂ú‡∂´‡∂±‡∂∫ ‡∂ö‡∂ª‡∂±‡∑Ä‡∑è.

#Negative ‡∑É‡∑Ñ zero values ignore ‡∂ö‡∂ª‡∂±‡∑Ä‡∑è.
```

```python
def arbitary_keyword_arguments(**kwargs):
    print("Keyword arguments are:", kwargs)
    print("Type of kwargs is:", type(kwargs))

arbitary_keyword_arguments(name="Alice", age=30, city="New York")

##**kwargs ‡∂ö‡∑í‡∂∫‡∂±‡∑ä‡∂±‡∑ö function ‡∂ë‡∂ö‡∂ß keyword arguments (key=value format)
#‡∂∏‡∑ú‡∂± ‡∂ú‡∂´‡∂±‡∂ö‡∑Ä‡∂≠‡∑ä pass ‡∂ö‡∂ª‡∂±‡∑ä‡∂± ‡∂Ø‡∑ô‡∂± mechanism ‡∂ë‡∂ö.
#‡∂∏‡∑ö ‡∑Ä‡∂ú‡∑ö arguments function ‡∂ë‡∂ö‡∂ß pass ‡∂ö‡∂ª‡∂±‡∂ö‡∑ú‡∂ß **kwargs ‡∂í‡∑Ä‡∑è dictionary ‡∂ë‡∂ö‡∂ö‡∑ä ‡∑Ä‡∑í‡∂Ø‡∑í‡∑Ñ‡∂ß receive ‡∂ö‡∂ª‡∂±‡∑Ä‡∑è.
```

```python
def arbitary_keyword_arguments(**kwargs):

    for key in kwargs.keys():
        print(key)

    print("==================================================")

    for value in kwargs.values():
        print(value)

    print("==================================================")

    for key, value in kwargs.items():
        print(f"{key} = {value}")


arbitary_keyword_arguments(name="Alice", age=30, city="New York")

#‡∂∏‡∑ô‡∂≠‡∂± code ‡∂ë‡∂ö ‡∂ö‡∂ª‡∂±‡∑ä‡∂±‡∑ö **kwargs (keyword arguments) dictionary ‡∂ë‡∂ö loop ‡∂ö‡∂ª‡∂Ω‡∑è
#keys, values, ‡∑É‡∑Ñ key-value pairs ‡∑Ä‡∑ô‡∂± ‡∑Ä‡∑ô‡∂±‡∂∏ print ‡∂ö‡∑í‡∂ª‡∑ì‡∂∏
```

```python
def avarage_score(**score):
    if len(score) == 0:
        print("No subjects provided.")
        return  # exit the function to avoid division by zero

    total = 0
    count = 0

    for subject, marks in score.items():
        total += marks
        count += 1
        print(f"{subject}: {marks}")

    avg = total / count
    print("=====================================")
    print("Average score is:", avg)


avarage_score()  # No subjects
avarage_score(Math=80, English=75, Science=90)  # With subjects
```
